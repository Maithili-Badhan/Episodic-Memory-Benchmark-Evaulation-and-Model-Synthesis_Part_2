```python
import shutil, os

# Path to the nested redundant directory
nested_path = "episodic-memory-benchmark/epbench/data/episodic-memory-benchmark/epbench/data"

# Path to the correct target location
target_path = "episodic-memory-benchmark/epbench/data"

# Move all contents up
for item in os.listdir(nested_path):
    s = os.path.join(nested_path, item)
    d = os.path.join(target_path, item)
    if os.path.exists(d):
        print(f"Skipping {item}, already exists at target")
    else:
        shutil.move(s, d)

# Remove the now-empty redundant folders
shutil.rmtree("episodic-memory-benchmark/epbench/data/episodic-memory-benchmark")
```

```python
from nbclient import NotebookClient
from nbformat import read, write

notebook_path = "/content/episodic-memory-benchmark/epbench/experiments/step_1_generation.ipynb"

with open(notebook_path) as f:
    nb = read(f, as_version=4)

client = NotebookClient(nb, timeout=600, kernel_name='python3')
client.execute()

# Save executed notebook
executed_path = "/content/ExecutedNotebook.ipynb"
with open(executed_path, 'w') as f:
    write(nb, f)

print("Notebook executed successfully:", executed_path)
```
